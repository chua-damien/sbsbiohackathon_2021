```python
def ENA_url(RunID):
  dir2 = ""
  if 9 < len(RunID) <= 12:
      dir2 = "0"*(12 - len(RunID)) + RunID[-(len(RunID)-9):] + "/"
  dirs = RunID[:6] + "/" + dir2 + RunID
  
  return "ftp://ftp.sra.ebi.ac.uk/vol1/fastq/" + dirs + "/" + RunID + "_1.fastq.gz", "ftp://ftp.sra.ebi.ac.uk/vol1/fastq/" + dirs + "/" + RunID + ".fastq.gz"

def get_RNA_seq_data(RunID, indexName):
  print('Getting expression values for ', RunID)
  paired, unpaired = ENA_url(RunID)  
  go = 0
  if RunID not in os.listdir(path): #If folder does not exist, set go to 1
    go = 1
  else:
    print(RunID, ' has already been done')

  if go == 1: 
    for i in range(3): 
        print('Attempt: ',i)

        !kallisto quant -i $indexName -o $RunID --single -l 200 -s 20 -t 2 <(curl -L -r 0-1000000000 -m 600 --speed-limit 1000000 --speed-time 30 $unpaired 2> $RunID+'.log')
        if '"n_processed": 0' in open(RunID+'/run_info.json','r').read():
            !kallisto quant -i $indexName -o $RunID --single -l 200 -s 20 -t 2 <(curl -L -r 0-1000000000 -m 600 --speed-limit 1000000 --speed-time 30  $paired 2> $RunID+'.log')
            
        if 'curl: (28)' not in open(RunID+'+.log','r').read(): 
            print('All OK! ')
            os.remove(RunID+'+.log')
            break

        if RunID+'+.log' in os.listdir(path):
            print('Download too slow. Will try ', 2-i, ' more times')
            os.remove(RunID+'+.log')
```

```python
#Compiling GO and gene descriptions from interproscan
s = open("/content/analysis_results (1).tsv", "r").readlines()
geneanno = open("c_jejuni_GOterms.txt", "w")

all = {}
allgenes = []
for i in s:
  total = i.split("\n")[0].split("\t")
  gene = total[0]
  allgenes.append(gene)
  if gene not in all:
    all[gene] = total[-1]
  else:
    if all[gene] != total[-1]:
      if total[-1] != "-":
        all[gene] += str("," + total[-1])

print(len(set(allgenes)))
print(len(all))

for i in all:
  geneanno.writelines(i + "\t" + all[i]+ "\n")
geneanno.close()
```
